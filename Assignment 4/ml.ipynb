```
from google.colab import drive
drive.mount('/content/drive')

import gzip
import _pickle
import random
from keras.models import Sequential
from keras.layers import Dense, MaxPooling2D,Conv2D, Flatten,Dropout
import pandas as pd
```

```
import os
import numpy as np
from sklearn.utils import shuffle
train_x = []
train_y = []
def get_train_data():
  global train_x,train_y
  lst = sorted(os.listdir('/content/train_pickle'))
  fold = random.sample(lst,1)
  #print(str(fold[0]))
  with gzip.open('/content/train_pickle/'+str(fold[0]),'rb') as file:
    arr = _pickle.load(file)
  m = len(arr[0])
  indx = random.sample(range(m),int(0.6*m))
  train_x0 = [arr[0][i] for i in indx]
  train_x0 = np.array(train_x0)
  op = [1,0]*int(0.6*m)
  op = np.array(op)
  m = op.shape
  #print(m)
  train_y0 = op.reshape(int(m[0]/2),2)
  #train_x = np.array(arr[0])
  #train_y = np.array([0]*m)
  m = len(arr[1])
  indx = random.sample(range(m),int(0.4*m))
  train_x1 = [arr[1][i] for i in indx]
  train_x1 = np.array(train_x1)
  op = [0,1]*int(0.4*m)
  op = np.array(op)
  m = op.shape
  #print(m)
  train_y1 = op.reshape(int(m[0]/2),2)
  #train_y1 = np.array([0,1]*int(0.4*m))
  #train_x1 = np.array(arr[1])
  #train_y1 = np.array([1]*m)
  
  train_y0 = train_y0.reshape(-1,2)
  train_y1 = train_y1.reshape(-1,2)
  train_x = np.vstack((train_x0,train_x1))
  train_y = np.vstack((train_y0,train_y1))
  #train_x = np.append(train_x,train_y,axis=1)
  #np.random.shuffle(train_x)
  #print(train_x.shape)
  #print(train_y.shape)
  train_x, train_y = shuffle(train_x,train_y,random_state=0)
  
  #train_y = train_x[:,-1]
  #train_x = train_x[:,0:-2]
  #print(train_x.shape)
  #print(train_y.shape)
  return train_x/255,train_y
```

```
model = Sequential()

#add model layers
model.add(Conv2D(32, kernel_size=3,strides=(2,2),activation='relu', input_shape=(210,160,5)))
model.add(MaxPooling2D(pool_size=(2,2),strides=(2,2)))
model.add(Conv2D(64, kernel_size=3,strides=(2,2),activation='relu'))
#model.add(Dropout(0.5))
model.add(MaxPooling2D(pool_size=(2,2),strides=(2,2)))
model.add(Flatten())
model.add(Dense(2048, activation='relu'))
#model.add(Dropout(0.5))
model.add(Dense(2,activation='softmax'))
#compile model using accuracy to measure model performance
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```


for epoch in range(10):
  train_x,train_y = get_train_data()
  print(train_x.shape)
  model.fit(train_x, train_y, epochs=epoch+1,initial_epoch=epoch,shuffle=True,batch_size=50)
  del train_x
  del train_y
```
